### Contexte :
L’agence AIL3C (Agence Indépendante de Lutte Contre la CyberCriminalité) a enfin réussi à localiser après plusieurs mois d’enquête la plus grande organisation de cyberpirates du web. Dans le but de récolter des informations compromettantes sur cette organisation, L’AIL3C a envoyé l’agent K57 au cœur de la base secrète ennemie. Cependant, un imprévu mit en péril l'opération d’infiltration, et l’agent K57 se retrouva bloqué dans une salle de conférence privée d’accès réseau et dont la composition des murs empêchait la communication vers l’extérieur via des dispositifs utilisant des ondes.
Malheureusement, l’agent K57 finit par se faire démasquer quelques heures plus tard. Il est donc impératif que cette situation ne se reproduise plus, et que notre département de recherche développe des solutions techniques permettant de contrer ce cas de figure.

### Problématique :
Comment l’agent peut-il arriver à communiquer avec l’agence sachant que la pièce dans laquelle il se trouve n’est pas équipée d’accés réseaux et dont les paroies sont hermétiques aux ondes ?

### Matériel mis à disposition :
L’agent n’avait à sa disposition qu’un micro branché au système d’audioconférence uniquement connecté au réseau lors de l’utilisation de la salle

### Piste de solution :
Pour remédier à ce problème nous avons pensé à utiliser le signal électrique produit par le mciro. En effet ce signal est en binaire et à l'aide du code ASCII nous pouvons lire ce signal sur 8 bits puis reproduire tous les caractères nécessaires pour former des phrases qui permettront aux agents de communiquer entre eux.

Voici le fonctionnement d’un microphone, l’onde sonore arrivant en entrée est convertie en signal électrique.


En bleu, le signal électrique créé par le microphone et en rouge, celui pouvant être interprété en 0 ou 1.



Grâce à nos combinaisons de 0 et de 1, nous pouvons à présent interpréter un message et donc communiquer avec l’agence.

### Exemple :
Pour illustrer notre solution si l'agent a besoin d'envoyer le caractère "e" à l'agent en charge des communications, il n'a qu'à envoyer le signal binaire "1100101" et l'agent qui reçoit le message pourra le décrypter et le lire.
Cela permettrait, dans le cas où une situation similaire au cas de l'agent K57 venait à se reproduire, à l'agent sur le terrain d'envoyer rapidement un message de demande d'aide et ainsi de lui sauver la vie.
